#### Cleaning the social network (239 users)

```python
import numpy as np
import pandas as pd
```

```python
adj_mat = pd.read_csv("./adj_blob.csv",index_col=0)
```

```python
adj_mat
```

```python
# convert the adjacency matrix to scipy sparse matrix
from scipy import sparse
adj_mat_sparse = sparse.csr_matrix(adj_mat.values)
```

```python
adj_mat_sparse
```

#### Converting tweet-level  labels to user-level labels

```python
data_mat = pd.read_csv("./data_blob.csv",index_col=0)
```

```python
data_mat
```

```python
data_mat.columns
```

```python
labels_tweet = data_mat['label'].values
```

```python
usernames_tweet = data_mat['screenname'].values
```

```python
len(usernames_tweet)
```

```python
len(np.unique(usernames_tweet))
```

```python
# take the most frequent tweet labels as the label for that user
user_labels = data_mat.groupby(['screenname'])['label'].apply(lambda x: x.value_counts().index[0]).reset_index()
```

```python
user_labels
```

#### Aggregating tweet-level  features (5-gram) to user-level features

```python
# Process the node features
X_df = pd.read_csv("./mat_5gram_blob.csv",index_col=0)
```

```python
X_df
```

```python
node_features = pd.pivot_table(X_df, index=['screenname'],values=X_df.columns,aggfunc=np.sum)
```

```python
node_features
```

```python
user_labels.index = user_labels['screenname']
```

```python
# reorder the nodes by the order in the ajacency matrix
ordered_nodes = np.array(adj_mat.index)
ordered_adj_mat = adj_mat_sparse
ordered_node_labels = user_labels.loc[ordered_nodes]['label'].values
ordered_node_features = node_features.loc[ordered_nodes].values
```

```python
# output the data
np.savez("5_gram_processed_mats.npz",ordered_nodes,ordered_adj_mat,ordered_node_labels,ordered_node_features)
```
